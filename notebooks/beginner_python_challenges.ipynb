**Learning objectives:**

````python
def is_even(n: int) -> bool:
    """
    Return True if n is even, False otherwise.
    """
    return n % 2 ==0

# Test cases
start = time.time()
try:
    assert is_even(4) == True
    assert is_even(7) == False
    assert is_even(0) == True
    assert is_even(-2) == True
    assert is_even(-3) == False
    elapsed = time.time() - start
    print(f"âœ“ All tests passed! ({elapsed*1000:.2f}ms)")
except AssertionError as e:
    print(f"âœ— Test failed: {e}")
except Exception as e:
    print(f"âœ— Error: {e}")
````

## Challenge 06: Count Vowels

Write a function that counts the number of vowels in a string.

**Learning objectives:**
- String iteration
- Membership testing (in operator)
- Case-insensitive comparison

````python
def count_vowels(text: str) -> int:
    """
    Return the number of vowels (a, e, i, o, u) in text.
    Case-insensitive.
    """
    return count_vowels_helper(text)

def count_vowels_helper(text):
    vowels = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count
# Test cases
start = time.time()
try:
    assert count_vowels("hello") == 2
    assert count_vowels("AEIOU") == 5
    assert count_vowels("xyz") == 0
    assert count_vowels("") == 0
    assert count_vowels("Programming") == 3
    elapsed = time.time() - start
    print(f"âœ“ All tests passed! ({elapsed*1000:.2f}ms)")
except AssertionError as e:
    print(f"âœ— Test failed: {e}")
except Exception as e:
    print(f"âœ— Error: {e}")
````

## Challenge 07: Reverse String

Write a function that reverses a string.

**Learning objectives:**
- String slicing
- String manipulation
- Pythonic patterns

````python
def reverse_string(text: str) -> str:
    """
    Return the reversed string.
    """
    return reverse_string_helper(text)

def reverse_string_helper(text):
    return text[::-1]

# Test cases
start = time.time()
try:
    assert reverse_string("hello") == "olleh"
    assert reverse_string("Python") == "nohtyP"
    assert reverse_string("") == ""
    assert reverse_string("a") == "a"
    assert reverse_string("12345") == "54321"
    elapsed = time.time() - start
    print(f"âœ“ All tests passed! ({elapsed*1000:.2f}ms)")
except AssertionError as e:
    print(f"âœ— Test failed: {e}")
except Exception as e:
    print(f"âœ— Error: {e}")
````

## Challenge 08: Filter Evens

Write a function that filters and returns only even numbers from a list.

**Learning objectives:**
- List comprehension
- Filtering patterns
- Creating new lists

````python
def filter_evens(numbers: list[int]) -> list[int]:
    """
    Return a list containing only the even numbers.
    """
    return [num for num in numbers if num % 2 ==0]
# Test cases
start = time.time()
try:
    assert filter_evens([1, 2, 3, 4, 5, 6]) == [2, 4, 6]
    assert filter_evens([1, 3, 5, 7]) == []
    assert filter_evens([2, 4, 6, 8]) == [2, 4, 6, 8]
    assert filter_evens([]) == []
    assert filter_evens([-2, -1, 0, 1, 2]) == [-2, 0, 2]
    elapsed = time.time() - start
    print(f"âœ“ All tests passed! ({elapsed*1000:.2f}ms)")
except AssertionError as e:
    print(f"âœ— Test failed: {e}")
except Exception as e:
    print(f"âœ— Error: {e}")
````

## Challenge 10: Safe Division

Write a function that safely divides two numbers, returning 0 if division by zero.

**Learning objectives:**
- Exception handling
- try-except blocks
- Defensive programming

````python
def safe_divide(a: float, b: float) -> float:
    """
    Return a/b, or 0 if b is zero.
    """
    if b == 0:
        return 0
    return a / b

# Test cases
start = time.time()
try:
    assert safe_divide(10, 2) == 5.0
    assert safe_divide(10, 3) == 10/3
    assert safe_divide(10, 0) == 0
    assert safe_divide(0, 5) == 0
    assert safe_divide(-10, 2) == -5.0
    elapsed = time.time() - start
    print(f"âœ“ All tests passed! ({elapsed*1000:.2f}ms)")
except AssertionError as e:
    print(f"âœ— Test failed: {e}")
except Exception as e:
    print(f"âœ— Error: {e}")
````
